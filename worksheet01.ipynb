```python
#Task3
import numpy as np

x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])
```

```python
#1:
# Adding the two arrays x and y
add_result = x + y
print("Addition of x and y:\n", add_result)

```

```python
# Subtracting the two arrays x and y
subtract_result = x - y
print("Subtraction of x and y:\n", subtract_result)
```

```python
# Multiplying array x with an integer, for example, 3
multiply_result = x * 3
print("Multiplication of x with 3:\n", multiply_result)
```

```python
# Squaring each element of array x
square_result = np.square(x)
print("Square of each element of x:\n", square_result)
```

```python
# Dot product of v and w
dot_vw = np.dot(v, w)
print("Dot product of v and w:", dot_vw)

# Dot product of x and v
dot_xv = np.dot(x, v)
print("Dot product of x and v:\n", dot_xv)

# Dot product of x and y
dot_xy = np.dot(x, y)
print("Dot product of x and y:\n", dot_xy)
```

```python
# Concatenate x and y along row (axis 1)
concat_xy_row = np.concatenate((x, y), axis=1)
print("Concatenation of x and y along row:\n", concat_xy_row)

# Concatenate v and w along column (axis 0)
concat_vw_column = np.concatenate((v[:, np.newaxis], w[:, np.newaxis]), axis=1)
print("Concatenation of v and w along column:\n", concat_vw_column)
```

```python
# Trying to concatenate x and v
try:
    concat_xv = np.concatenate((x, v), axis=1)
    print("Concatenation of x and v:\n", concat_xv)
except ValueError as e:
    print("Error:", e)
```

```python
#Task4
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

```

```python
#1. Prove A.A−1 = I.
# Compute the inverse of A
A_inv = np.linalg.inv(A)

# Compute A * A_inv
identity_matrix = np.dot(A, A_inv)
print("A * A^-1 = \n", identity_matrix)

```

```python
#2. Prove AB ̸= BA.
# Compute AB and BA
AB = np.dot(A, B)
BA = np.dot(B, A)

# Print the results
print("AB = \n", AB)
print("BA = \n", BA)
```

```python
#3
# Compute (AB)^T
AB_T = np.transpose(AB)

# Compute B^T * A^T
BT_AT = np.dot(np.transpose(B), np.transpose(A))

# Print the results
print("Transpose of AB = \n", AB_T)
print("B^T * A^T = \n", BT_AT)

```

```python
#4.
# Coefficient matrix A
A = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])

# Constants vector B
B = np.array([-1, -3, 9])

# Compute the inverse of A
A_inv = np.linalg.inv(A)

# Solve for X
X = np.dot(A_inv, B)
print("Solution X = \n", X)
```
