```python
low_cutoff = df[['battery_power', 'ram', 'n_cores']].quantile(0.4)  # 40th percentile
mid_cutoff = df[['battery_power', 'ram', 'n_cores']].quantile(0.7)  # 70th percentile

def classify_performance(row):
    if (row['ram'] > mid_cutoff['ram'] and
        row['battery_power'] > mid_cutoff['battery_power'] and
        row['n_cores'] > mid_cutoff['n_cores']):
        return 'High'
    elif (row['ram'] > low_cutoff['ram'] and
          row['battery_power'] > low_cutoff['battery_power'] and
          row['n_cores'] > low_cutoff['n_cores']):
        return 'Mid'
    else:
        return 'Low'

df['performance_tier'] = df.apply(classify_performance, axis=1)
```

```python
import scipy.stats as stats
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
def detect_outliers_zscore(data, feature, threshold=3):
  z_scores = stats.zscore(data[feature])
  outliers = data[(abs(z_scores) > threshold)][feature]
  inliers = data[(abs(z_scores) <= threshold)][feature]
  print(f"Feature: {feature}")
  print(f"Number of outliers: {len(outliers)}")
  print(f"Examples of outliers: {outliers.head()}")  # View top 5 outliers
  print(f"Number of inliers: {len(inliers)}")
# Apply z-score outlier detection to each numerical feature
for feature in numerical_features:
    detect_outliers_zscore(df.copy(), feature)

threshold = 3
for feature in numerical_features:
    z_scores = stats.zscore(df[feature])
    outliers = df[(abs(z_scores) > threshold)][feature]
    inliers = df[(abs(z_scores) <= threshold)][feature]
    plt.boxplot(df[feature], notch=True, vert=True, patch_artist=True)
    plt.scatter(df.index[outliers], outliers, color='red', label='Outliers')
    plt.xlabel(feature)
    plt.ylabel('Value')
    plt.title(f"Box Plot with Outliers for {feature}")
    plt.legend()
    plt.show()
    plt.close()
```

Building Primary Model

```python
import numpy as np

class MulticlassLogisticRegression:
    def __init__(self, lr=0.01, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None
        self.losses = []

    def _softmax(self, z):
        exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
        return exp_z / np.sum(exp_z, axis=1, keepdims=True)

    def fit(self, X, y):
        n_samples, n_features = X.shape
        n_classes = len(np.unique(y))

        y_one_hot = np.eye(n_classes)[y]
        self.weights = np.zeros((n_features, n_classes))
        self.bias = np.zeros((1, n_classes))

        for _ in range(self.n_iters):
            linear_pred = np.dot(X, self.weights) + self.bias
            probs = self._softmax(linear_pred)

            loss = -np.mean(np.sum(y_one_hot * np.log(probs + 1e-9), axis=1))
            self.losses.append(loss)

            dw = (1 / n_samples) * np.dot(X.T, (probs - y_one_hot))
            db = (1 / n_samples) * np.sum(probs - y_one_hot, axis=0, keepdims=True)

            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    def predict(self, X):
        linear_pred = np.dot(X, self.weights) + self.bias
        probs = self._softmax(linear_pred)
        return np.argmax(probs, axis=1)
```

```python
lr=MulticlassLogisticRegression()
lr.fit(X_train,y_train)
y_pred_lr=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report,precision_score, recall_score, f1_score, confusion_matrix
print(f"Accuracy Score = {accuracy_score(y_test,y_pred_lr)}")
print(f"precission Score = {precision_score(y_test,y_pred_lr,average='macro')}")
print(f"Recall Score = {recall_score(y_test,y_pred_lr,average='macro')}")
print(f"F1 Score = {f1_score(y_test,y_pred_lr,average='macro')}")
```

The model's accuracy is 0.94 which is good but precession and recall is low
